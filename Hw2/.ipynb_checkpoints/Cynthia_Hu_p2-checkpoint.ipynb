Whereas with digit classification, the input is relatively dense: a 28x28 matrix of pixels, many of which are non-zero, here we'll represent each document with a "bag-of-words" model.
```python
categories = ['alt.atheism', 'talk.religion.misc', 'comp.graphics', 'sci.space']
newsgroups_train = fetch_20newsgroups(subset='train',
                                      remove=('headers', 'footers', 'quotes'),
                                      categories=categories)
newsgroups_test = fetch_20newsgroups(subset='test',
                                     remove=('headers', 'footers', 'quotes'),
                                     categories=categories)

num_test = len(newsgroups_test.target)
test_data, test_labels = newsgroups_test.data[int(num_test/2):], newsgroups_test.target[int(num_test/2):]
dev_data, dev_labels = newsgroups_test.data[:int(num_test/2)], newsgroups_test.target[:int(num_test/2)]
train_data, train_labels = newsgroups_train.data, newsgroups_train.target

print('training label shape:', train_labels.shape)
print('test label shape:', test_labels.shape)
print('dev label shape:', dev_labels.shape)
print('labels names:', newsgroups_train.target_names)

#print train_data[:5]
```

```python
def P2():
    vectorizer = CountVectorizer(min_df=1)
    # tokenize and count the occurrences of the token in each message in the train_data
    X = vectorizer.fit_transform(train_data)
    # X.toarray()

    # a.check the size of the vector feature matrix
    print("a. The size of the vocabulary is %d." % len(vectorizer.vocabulary_))
    # nnz - Get the count of explicitly-stored values (nonzeros) from a sparse matrix
    print("a. The average number of non-zero features per example is %.2f." % (X.nnz * 1.0 / X.shape[0]))
    print("a. %.2f%% of the entries in the matrix are non-zero." % (X.nnz * 100.0 / (X.shape[0] * X.shape[1])))
    print()

    # b. check specific features
    feature = vectorizer.get_feature_names()
    print("b. The first feature is %s and the last feature is %s." % (feature[0], feature[len(feature)-1]))
    print()

    # c. use own vocabulary
    my_vocabulary = ["atheism", "graphics", "space", "religion"]
    vectorizer2 = CountVectorizer(min_df=1, vocabulary=my_vocabulary)
    X2 = vectorizer2.fit_transform(train_data)
    print("c. The size of the training vector is ", X2.shape)
    print("c. The customized vocabulary is: ", vectorizer2.vocabulary_)
    print("c. The average number of non-zero features per example is %.2f." % (X2.nnz * 1.0 / X2.shape[0]))
    print()

    # d. bi-gram and tri-gram character features
    n_vectorizer = CountVectorizer(min_df=1, analyzer='char', ngram_range=(2,3))
    X_n = n_vectorizer.fit_transform(train_data)
    print("d. The size of vocabulary for bi-gram and tri-gram character features is %d." % len(n_vectorizer.vocabulary_))
    print()

    # e. prune words appear in fewer than 10 documents
    pr_vectorizer = CountVectorizer(min_df=10) # use min_df=10
    X_pr = pr_vectorizer.fit_transform(train_data)
    print("e. The size of vocabulary after pruning words is %d." % len(pr_vectorizer.vocabulary_))
    print()

    # f. fraction of missing vocabulary
    X_train = vectorizer.fit_transform(train_data)
    X_dev = vectorizer.fit_transform(dev_data)
    print("f. %.2f%% of the words in dev data are missing from the vocabulary." 
          % ((X_dev.shape[1]-X_train.shape[1]) * 1.0 / X_train.shape[1]))

P2()
```

(3) Use the default CountVectorizer options and report the f1 score (use metrics.f1_score) for a k nearest neighbors classifier; find the optimal value for k. Also fit a Multinomial Naive Bayes model and find the optimal value for alpha. Finally, fit a logistic regression model and find the optimal value for the regularization strength C using l2 regularization. A few questions:

a. Why doesn't nearest neighbors work well for this problem?

b. Any ideas why logistic regression doesn't work as well as Naive Bayes?

c. Logistic regression estimates a weight vector for each class, which you can access with the coef\_ attribute. Output the sum of the squared weight values for each class for each setting of the C parameter. Briefly explain the relationship between the sum and the value of C.

```python
def P3():
    # shall I use pipeline here?

    # create feature vector using default CountVectorizer options
    vectorizer = CountVectorizer()
    vectorizer.fit(train_data)
    # transform both train_data and test_data so that both have the same features
    train_X = vectorizer.transform(train_data)
    test_X = vectorizer.transform(test_data)

    # k nearest neighbor
    k_values ={'n_neighbors': [1,3,5,7,9,11]}
    knn = KNeighborsClassifier()
    gs_knn = GridSearchCV(knn, k_values)
    gs_knn.fit(train_X,train_labels)
    print('The best k value for K Nearest Neighbor model is ', gs_knn.best_params_)
    knn_best = gs_knn.best_estimator_
    knn_best.fit(train_X,train_labels)
    knn_pred = knn_best.predict(test_X)
    print('Test accuracy for K Nearest Neighbor model is : %.3f' 
          % metrics.f1_score(test_labels, knn_pred, average='weighted'))
    print() # 0.415

    # multinomial Naive Bayes
    alphas = {'alpha': [0.001, 0.01, 0.1, 0.5, 1.0, 2.0, 10.0]}
    mnb = MultinomialNB(alpha=1.0)
    gs_mnb = GridSearchCV(mnb, alphas)
    gs_mnb.fit(train_X,train_labels)
    print('The best alpha for Multinomial Naive Bayes model is ', gs_mnb.best_params_)
    mnb_best = gs_mnb.best_estimator_
    mnb_pred = mnb_best.predict(test_X)
    print('Test accuracy for Multinomial Naive Bayes model is : %.3f' 
          % metrics.f1_score(test_labels, mnb_pred, average='weighted')) # 0.78
    print()

    # Logistic Regression
    C_values = {'C': [1.0, 2.0, 10.0]}
    lgr = LogisticRegression(penalty='l2')
    gs_lgr = GridSearchCV(lgr, C_values)
    gs_lgr.fit(train_X,train_labels)
    print('The best C value for Logistic Regression model is ', gs_lgr.best_params_)
    lgr_best = gs_lgr.best_estimator_
    lgr_pred = lgr_best.predict(test_X)
    print('Test accuracy for Logistic Regression model is : %.3f' 
          % metrics.f1_score(test_labels, lgr_pred, average='weighted'))

    # explore the relation between C value and the sum of squared weight value for each class
    C_list = list(C_values.values())[0]
    for c in C_list:
        lgr2 = LogisticRegression(penalty='l2', C=c)
        lgr2.fit(train_X,train_labels)
        a = lgr2.coef_**2
        print('C value is ',c)
        print('The sum of squared weight value for each class is \n', a.sum(axis=1))

P3()
```

ANSWER:

a. As the feature vector is a sparse matrix, it's difficult to find similar neighbors. Thus, nearest neighbor model doesn't work well in this case.

b. We use L2 regulation for logistic regression model, so there are still too many features. Too many features may lead to overfitting issues.

c. C value and the sum of squared weight value for each class is positively related. The higher the C value, the higher the sum of squared weight.


(4) Train a logistic regression model. Find the 5 features with the largest weights for each label -- 20 features in total.
```python
def P4():
    # unigram word feature
    # create feature vector for unigram word feature
    vectorizer = CountVectorizer(stop_words='english')
    vectorizer.fit(train_data)
    feature_ls = vectorizer.get_feature_names()
    # tranform both train_data and test_data
    train_X = vectorizer.transform(train_data)
    test_X = vectorizer.transform(test_data)
    # fit logistic regression model
    lgr = LogisticRegression(penalty='l2', C=1.0)
    lgr.fit(train_X,train_labels)
    coefficient = lgr.coef_

#     find top 5 features for each class
    label_ls = newsgroups_train.target_names
    class_top_feature = {}
    print("Top 20 weight table for Unigram features: ")
    print(' '*15,label_ls)
    for i in range(4): # loop through each class
        coef_list = list(coefficient[i])  
        coef_sorted =sorted(coef_list,reverse=True)[:5] # for each class, sort the weight and pick top 5 weight
        top_feature=[]
        top_weight = []
        # for the top 5 weights, look for the name of the features
        for j in range(5):
            loc=coef_list.index(coef_sorted[j])
            feature = feature_ls[loc]
            top_feature.append(feature)
            weight = coefficient.transpose()[loc]
            top_weight.append(weight)     
    #         print(feature_ls[loc], coefficient.transpose()[loc])       
            print("{:<15} {:<15.2f} {:<15.2f} {:<15.2f} {:<15.2f}".format(feature, weight[0], weight[1], weight[2], weight[3]))
        class_top_feature[label_ls[i]] = top_feature
    print()

    # bigram word feature
    # create feature vector for bigram word feature
    vectorizer2 = CountVectorizer(ngram_range=(2,2) ,stop_words='english')
    vectorizer2.fit(train_data)
    feature_ls2 = vectorizer2.get_feature_names()
    # tranform both train_data and test_data
    train_X2 = vectorizer2.transform(train_data)
    test_X2 = vectorizer2.transform(test_data)
    # fit logistic regression model
    lgr2 = LogisticRegression(penalty='l2', C=1.0)
    lgr2.fit(train_X2,train_labels)
    coefficient2 = lgr2.coef_

    class_top_feature2 = {}
    print("Top 20 weight table for Bigram features: ")
    print(' '*20,label_ls)
    for i in range(4): # loop through each class
        coef2_list = list(coefficient2[i])  
        coef2_sorted =sorted(coef2_list,reverse=True)[:5] # for each class, sort the weight and pick top 5 weight
        top_feature2=[]
        top_weight2 = []
        # for the top 5 weights, look for the name of the features
        for j in range(5):
            loc2=coef2_list.index(coef2_sorted[j])
            feature2 = feature_ls2[loc2]
            top_feature2.append(feature)
            weight2 = coefficient2.transpose()[loc2]
            top_weight2.append(weight2)           
            print("{:<20} {:<15.2f} {:<15.2f} {:<15.2f} {:<15.2f}".format(feature2, weight2[0], weight2[1], weight2[2], weight2[3]))
        class_top_feature2[label_ls[i]] = top_feature2

P4()
```

(5) Try to improve the logistic regression classifier by passing a custom preprocessor to CountVectorizer. The preprocessing function runs on the raw text, before it is split into words by the tokenizer. Your preprocessor should try to normalize the input in various ways to improve generalization. For example, try lowercasing everything, replacing sequences of numbers with a single token, removing various other non-letter characters, and shortening long words. If you're not already familiar with regular expressions for manipulating strings, see https://docs.python.org/2/library/re.html, and re.sub() in particular. With your new preprocessor, how much did you reduce the size of the dictionary?

For reference, I was able to improve dev F1 by 2 points.

```python
import re

def empty_preprocessor(s):
    return s

def default_preprocessor(x): 
    one = x.lower()
    # remove digit and letter combination
    two = re.sub(r'\w*[0-9]\w*', '<NUM>', one)
    # remove some non-letter characters
    three = re.sub('[\_\#\-]+', "", two)
    # detect email address domain
    four = re.sub("@[\w.]+", '<Email>', three) 
    five = four.replace('<NUM><Email>', '<Email>') 
    return ' '.join(five.strip().split())

def P5():
    # from P2 and P3 we know that baseline: 26879 features and 0.734 test accuracy (C=1.0)
    # vectorizer = CountVectorizer(preprocessor=empty_preprocessor,stop_words='english')
    # besides using the customized preprocessor, also remove common english words from feature extraction
    vectorizer = CountVectorizer(preprocessor=default_preprocessor,stop_words='english')
    vectorizer.fit(train_data)
    train_X = vectorizer.transform(train_data)
    test_X = vectorizer.transform(test_data)
    print("The size of vocabulary is %d, reduced by %.2f%%." 
          % (len(vectorizer.vocabulary_), 1-len(vectorizer.vocabulary_)*1.0/26879))
    lgr = LogisticRegression(penalty='l2', C=1.0)
    lgr.fit(train_X,train_labels)
    lgr_pred = lgr.predict(test_X)
    print('Test accuracy with the new preprocessor is : %.3f'
          % metrics.f1_score(test_labels, lgr_pred, average='weighted'))

P5()
```

(6) The idea of regularization is to avoid learning very large weights (which are likely to fit the training data, but not generalize well) by adding a penalty to the total size of the learned weights. That is, logistic regression seeks the set of weights that minimizes errors in the training data AND has a small size. The default regularization, L2, computes this size as the sum of the squared weights (see P3, above). L1 regularization computes this size as the sum of the absolute values of the weights. The result is that whereas L2 regularization makes all the weights relatively small, L1 regularization drives lots of the weights to 0, effectively removing unimportant features.

Train a logistic regression model using a "l1" penalty. Output the number of learned weights that are not equal to zero. How does this compare to the number of non-zero weights you get with "l2"? Now, reduce the size of the vocabulary by keeping only those features that have at least one non-zero weight and retrain a model using "l2".

Make a plot showing accuracy of the re-trained model vs. the vocabulary size you get when pruning unused features by adjusting the C parameter.

Note: The gradient descent code that trains the logistic regression model sometimes has trouble converging with extreme settings of the C parameter. Relax the convergence criteria by setting tol=.01 (the default is .0001).

```python
def P6():
    # Keep this random seed here to make comparison easier.
    np.random.seed(0)
    # Generate feature vector
    vectorizer = CountVectorizer(stop_words='english')
    vectorizer.fit(train_data)
    train_X = vectorizer.transform(train_data)
    test_X = vectorizer.transform(test_data)

    # fit logistic regression using L2 regulation as the baseline
    lgr_l2 = LogisticRegression(penalty='l2', C=1.0)
    lgr_l2.fit(train_X,train_labels)
    lgr_l2_pred = lgr_l2.predict(test_X)
    coef_l2 = lgr_l2.coef_  # shape (4,26577)
    # find how many features with at least one non-zero weight
    feature_l2_sum = abs(coef_l2).sum(axis=0)
    nonzero_weight_cnt = coef_l2[coef_l2 != 0].size
    print("Unpruned model with L2 regulation has %d featues and the accuracy is %.3f."
          % (feature_l2_sum[feature_l2_sum != 0].shape[0], metrics.f1_score(test_labels, lgr_l2_pred, average='weighted')))

    # pruning features using C parameter
    C_value = [0.1, 0.5, 1.0, 1.5, 2.0, 3.0, 4.0, 10.0]
    nonzero_weight_count = []
    accuracy = []
    vocabulary_size = []
    for i in C_value:
        # using model with L1 regulation to pruning features
        # and then fit model with L2 regulation
        lgr_l1 = LogisticRegression(penalty='l1', C=i)
        lgr_l1.fit(train_X,train_labels)
        lgr_l1_pred = lgr_l1.predict(test_X)
        coef_l1 = lgr_l1.coef_
        feature_l1_sum = abs(coef_l1).sum(axis=0)

        # build a new feature vector using smaller vocabulary
        big_vocabulary = np.array(vectorizer.get_feature_names())
        small_vocabulary = list(big_vocabulary[feature_l1_sum != 0])
        vectorizer2 = CountVectorizer(vocabulary=small_vocabulary, stop_words='english')
        vectorizer2.fit(train_data)
        train_X2 = vectorizer2.transform(train_data)
        test_X2 = vectorizer2.transform(test_data)

        # retain a model using L2 regulation and smaller vocabulary
        lgr_l2b = LogisticRegression(penalty='l2', C=i, tol=.01)
        lgr_l2b.fit(train_X2,train_labels)
        lgr_l2b_pred = lgr_l2b.predict(test_X2)
        coef_l2b = lgr_l2b.coef_
        feature_l2b_sum = abs(coef_l2b).sum(axis=0)
        accuracy.append(metrics.f1_score(test_labels, lgr_l2b_pred, average='weighted'))
        vocabulary_size.append(feature_l2b_sum[feature_l2b_sum != 0].shape)
        nonzero_weight_count.append(coef_l2b[coef_l2b != 0].size)
        print()
        print("Non-zero weight count: ", coef_l2b[coef_l2b != 0].size)
        print("Vocabulary Size: ", len(vectorizer2.vocabulary_ )) # the same as feature_l2b_sum[feature_l2b_sum != 0].shape
        print("Accuracy with %.1f value of C is %.3f."
              % (i, metrics.f1_score(test_labels, lgr_l2b_pred, average='weighted')))
    # print(accuracy)
    # print(size)
    size = [i[0] for i in vocabulary_size]
    plt.plot(size, accuracy)
    plt.title("Accuracy vs. Vocabulary Size")
    
P6()
```

(7) Use the TfidfVectorizer -- how is this different from the CountVectorizer? Train a logistic regression model with C=100.

Make predictions on the dev data and show the top 3 documents where the ratio R is largest, where R is:

maximum predicted probability / predicted probability of the correct label

What kinds of mistakes is the model making? Suggest a way to address one particular issue that you see.

```python
def P7():
    # create normalized feature vector
    tdvector = TfidfVectorizer() 
    tdvector.fit(train_data)
    # tranform train_data and dev_data
    train_td = tdvector.transform(train_data)
    dev_td = tdvector.transform(dev_data)

    # fit logistic regression model with C=100
    lgr_td = LogisticRegression(penalty='l2', C=100)
    lgr_td.fit(train_td,train_labels)
    lgr_td_pred = lgr_td.predict(dev_td)
    lgr_td_prob = lgr_td.predict_proba(dev_td) # probability for each document by each class
    print("Test Accuary is %.3f." % metrics.f1_score(dev_labels, lgr_td_pred, average='weighted')) 

    # Calculate R value for each document
    R_value = []
    # top 3 documents where the ratio R is largest means the most incorrect labeled document
    top_incorrect_doc = [] 
    for i in range(dev_td.shape[0]):
        max_pred_prob = lgr_td_prob[i].max()
        correct_label = dev_labels[i]
        prob_correct = lgr_td_prob[i][correct_label]
    #     print(lgr_td_prob[i])
    #     print("correct label: ", lgr_td_pred[i])
        R_value.append(max_pred_prob * 1.00 / prob_correct)

    R_sorted = sorted(R_value, reverse = True)
    # R_sorted[:5]
    print("\n Top 3 documents with the highest R ratio is: \n")
    for i in range(3):
        loc = R_value.index(R_sorted[i])
        top_incorrect_doc.append(dev_data[loc])
        print(i+1,". ", dev_data[loc])
P7()
```

ANSWER: TfidfVectorizer is equivalent to CountVectorizer followed by TfidfTransformer. TfidfTranformer applies Term Frequency Inverse Document Frequency normalization to a sparse matrix of occurrence counts.
From the top 3 incorrectly classified documents, I notice that the documents contain numbers and non-letter characters which may require preprocessing. Further as it's a sparse matrix with a lot of zeros, the differences between normalized feature weights are smaller. Reduce the number of features may help, especially those with many zero weights.
