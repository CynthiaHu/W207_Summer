## Dataset Introduction & Initial Preparation

The goal of this competition is to predict the price of a home given 79 variables as inputs. The dataset comes from surveying 79 details about homes in Ames, Iowa. A detailed description of each variable can be found in the data_description.txt file in the [data tab](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data) in the competition's page. 

We'll start by loading in our training and test datasets into memory. We'll split the training dataset into train and development subsets. The training dataset will be used for EDA and model training. The development dataset will be used for model evaluation & selection. We'll use the test dataset for evaluating the final model.

Let's import the modules and tools we'll use for the notebook, and load train & test datasets into pandas DataFrame:

```python
%matplotlib inline

# Data manipulation libraries
import pandas as pd
import numpy as np

# Plotting tools
import seaborn as sns
import matplotlib.pyplot as plt

# Preprocessing tools
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline

# Regression models
from sklearn.linear_model import LinearRegression, Ridge, Lasso, RandomizedLasso, ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error

# My own transformers based on the EDA
from transformers import *

# Remove warnings because they look messy.
import warnings
warnings.filterwarnings('ignore')
```

We may need to engineer new features based on a combination of those two columns in order to see a relationship with log_SalePrice.

We'll examine the subset of features in the top left cluster:

```python
clustered_features = [label.get_text() for label in cg.ax_heatmap.xaxis.get_majorticklabels()]
important_features = clustered_features[:21]
important_features
```

Let's remake the clustered correlation heatmap with just this subset of features:

Two of these features, YearRemodAdd and YearBuilt, are yearly. Let's subtract the minimum years from those so that their minimum values are zero:

```python
for feat in ['YearRemodAdd', 'YearBuilt']:
    train[feat + '_rescaled'] = train[feat] - train[feat].min()
    important_features.remove(feat)
    important_features.append(feat + '_rescaled')
```

Now let's look at scatter/regression plots of the features with the label to see the relationships:

```python
# Reshape the feature list into a 5x4 ndarray so we can iterate through it in a grid
important_features.remove('SalePrice')
important_features_arr = np.array(important_features).reshape(5,4) 

# We'll do a 5x4 grid of subplots since we have 20 features
f, axs = sns.plt.subplots(5,4)
f.set_size_inches(16,20)

for i in range(4):
    for j in range(5):
        sns.regplot(data=train, x=important_features_arr[j,i], y='SalePrice', ax=axs[j,i])
```

Observations:
* There are a few features where the slope is skewed by zero values of the feature. We may be able to benefit by creating slope dummies for those and including them in the model. These features are YearRemodAdd, WoodDeckSF, 2ndFlrSF, MaxVnrArea, TotalBsmtSF, BsmtUnfSF
* A few features have outliers that skew the fit. These features are: 1stFlrSF, LotArea, LotFrontage, TotalBsmtSF, and GrLivArea

To deal with the zero-values in certain features skewing the lienar fit, we'll craft new dummy features that have a value of 1 when the feature is 0, and is 0 otherwise:

```python
zeroval_features = [
    'YearRemodAdd_rescaled'
    , 'WoodDeckSF'
    , '2ndFlrSF'
    , 'MasVnrArea'
    , 'TotalBsmtSF'
    , 'BsmtUnfSF'
]

for feat in zeroval_features:
    train['{}_zero'.format(feat)] = 0
    train.ix[train[feat] == 0, '{}_zero'.format(feat)] = 1
```

Let's re-examine the scatter/regression plots of these features with the label, using the new dummy features as the hue:

```python
for i in range(len(zeroval_features)):
    sns.lmplot(data=train, y='SalePrice', x=zeroval_features[i], hue=zeroval_features[i] + '_zero')
```

Those look like much better fits. We'll include these new interaction dummy features in our final model.

Let's see if we can improve the linear fits of some of the other features by removing outliers. We'll do this by subtracting the mean and dividing by the standard deviation, and then we'll remove observations with absolute values greater than 4 (this is arbitrarily chosen. We'll find the best value for it later).

```python
# Fill in missing values in the LotFrontage column with zeros, because they're likely zero because they don't have a front lot.
# If they don't have a lot, the square footage of their front lot is zero. :)
train['LotFrontage'] = train['LotFrontage'].fillna(0)

# Initialize our standard scaler, and fit transform the features we previously identified
stdscaler = StandardScaler()
feats_to_rescale = ['1stFlrSF', 'LotArea', 'LotFrontage', 'TotalBsmtSF', 'GrLivArea']
transformed_feats = stdscaler.fit_transform(train[feats_to_rescale])

# Craft new column names, and then build a new temp DataFrame with these rescaled features
new_column_names = [f + '_zscore' for f in ['1stFlrSF', 'LotArea', 'LotFrontage', 'TotalBsmtSF', 'GrLivArea']]
train_transformed_feats = pd.DataFrame(transformed_feats, columns=new_column_names)

# Append the temp DataFrame with the train DataFrame
train = pd.concat([train, train_transformed_feats], axis=1)

# Create a new DataFrame with the observatins containing outliers filtered out
train_filtered = train[
      (train['1stFlrSF_zscore'].abs() <= 4)
    & (train['LotArea_zscore'].abs() <= 4)
    & (train['LotFrontage_zscore'].abs() <= 4)
    & (train['TotalBsmtSF_zscore'].abs() <= 4)
    & (train['GrLivArea_zscore'].abs() <= 4)
]
```

```python
print(train_filtered['LotArea'].std())
print(train_filtered['LotArea'].mean())
```

```python
rescaled_feats = new_column_names  # rescaled_feats is a better variable name

f, axs = sns.plt.subplots(len(feats_to_rescale),1)
f.set_size_inches(8,20)

for i in range(len(feats_to_rescale)):
    sns.regplot(data=train_filtered, x=feats_to_rescale[i], y='SalePrice', ax=axs[i])
```

Those look like better fits. We don't see any particular observation skewing the entire line. This filter will be useful for regression models that are sensitive to outliers, e.g. standard Linear Regression or Gradient Boosting Regression.

## Summary

We explored the training dataset to find which continuous and categorical features, and found a number of features that will (likely) be helpful in predicting SalePrice, and engineering a few new ones.

Before feeding the training data into a machine learning model, we'll perform these steps to transform the data into its optimal form:
1. Subset the training dataset to only the features we identified as predictive of SalePrice
2. Fill in missing values of certain continous feature columns (LotFrontage, MasVnrArea) to zero
3. Subtract the minimum from the yearly features (YearRemodAdd, YearBuilt) so that their minimum is zero
4. Create categorical features that indicate whether certain continuous features (e.g. YearRemodAdd) equal zero
5. Set values of 6 to 9 in the OverallCond feature equal to 5.
6. StandardScale the continuous features (subtract the mean, divide by standard deviation)
7. Remove outliers that have abs(z-score) greather than a certain threshold (Note: we'll have to disable this step when predicting the dev and test sets)
8. One-hot encode the categorical features
9. Run PCA on the categorical features

Some of these steps have hyperparameters that we'll tune when it comes time to fit the predictive model. We'll experiment with different z-scores in step 6, different numbers of components in step 8, and whether we drop outliers or perform PCA at all.

We'll create an sklearn Pipeline that performs all of these transformatations. This pipeline will be useful when training ML models.

# Data Preprocessing

Let's build the transformation pipeline. Since standard sklearn transformers can't handle pandas DataFrames out of the box, we'll create custom ones to wrangle the training DataFrame into the right shape.

```python
categorical_variables_to_ohe = [
      'MSZoning'
    , 'MSSubClass'
    , 'Street'
    , 'Alley'
    , 'LotShape'
    , 'LandContour'
    , 'Utilities'
    , 'LotConfig'
    , 'LandSlope'
    , 'Neighborhood'
    , 'Condition1'
    , 'Condition2'
    , 'BldgType'
    , 'HouseStyle'
    , 'RoofStyle'
    , 'RoofMatl'
    , 'Exterior1st'
    , 'Exterior2nd'
    , 'MasVnrType'
    , 'ExterQual'
    , 'ExterCond'
    , 'Foundation'
    , 'BsmtQual'
    , 'BsmtCond'
    , 'BsmtExposure'
    , 'BsmtFinType1'
    , 'BsmtFinType2'
    , 'Heating'
    , 'HeatingQC'
    , 'CentralAir'
    , 'Electrical'
    , 'KitchenQual'
    , 'Functional'
    , 'FireplaceQu'
    , 'GarageType'
    , 'GarageFinish'
    , 'GarageQual'
    , 'GarageCond'
    , 'PavedDrive'
    , 'PoolQC'
    , 'Fence'
    , 'SaleType'
    , 'SaleCondition'
    , 'YrSold'
    , 'BsmtHalfBath'
    , 'MoSold'
    , 'KitchenAbvGr'
    , 'MiscFeature'
]

remove_outlier_columns = [
     '1stFlrSF'
    , 'LotArea'
    , 'LotFrontage'
    , 'TotalBsmtSF'
    , 'GrLivArea'
]

cols_to_add_zero_indic_for = [
    'YearRemodAdd'
    , 'WoodDeckSF'
    , '2ndFlrSF'
    , 'MasVnrArea'
    , 'TotalBsmtSF'
    , 'BsmtUnfSF'
    , 'LotFrontage'
    , 'BsmtFinSF1'
    , 'ScreenPorch'
    , 'BsmtFinSF2'
    , 'EnclosedPorch'
]

columns_to_drop = [
      'Id'
    , '3SsnPorch'
    , 'PoolArea'
    , 'BsmtFinSF2'
    , 'LowQualFinSF'
    , 'MiscVal'
]
```

```python
transformation_steps = [
      ('drop_columns', ColumnDropper(columns_to_drop=columns_to_drop))
    , ('convert_categoricals_to_string', ConvertCategoricalsToString())
    , ('impute_missing', ImputeMissings())
    , ('subtract_mins', SubtractMinimum())
    , ('add_zero_indic', AddZeroIndicatorColumns())
#     , ('correct_overallcond', CorrectOverallCond())
    , ('stdscl_continuous', StandardScaleContinuousFeatures())
#     , ('remove_outliers', RemoveOutliers())
    , ('ohe_cat', OneHotEncodeCategoricals())
#     , ('extract_to_ndarray', DataFrameExtractor())
#     , ('poly_features', PolynomialFeatures(degree=2))
#     , ('gradient_boosting', GradientBoostingRegressor())
]

transformation_pipeline = Pipeline(steps=transformation_steps)
```

```python
transformation_pipeline.fit(train_orig)
train_transformed = transformation_pipeline.transform(train_orig)
```
