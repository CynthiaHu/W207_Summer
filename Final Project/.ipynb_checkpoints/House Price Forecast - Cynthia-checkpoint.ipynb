**Log transform skewed numerical variables**

```python
#log transform the target:
df1["SalePrice"] = np.log1p(df1["SalePrice"])

#log transform skewed numeric features:
# numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index

skewed_cols = df1[num_cols].apply(lambda x: skew(x.dropna())) #compute skewness
skewed_cols = skewed_cols[skewed_cols > 0.75]
skewed_cols = skewed_cols.index

df1[skewed_cols] = np.log1p(df1[skewed_cols])
```

** Convert Categorical Variables into Dummy Variables **

```python
#convert categorical variable into dummy
df1_dummy = pd.get_dummies(df1[cat_cols])
df1_dummy.head()
```

** Apply PCA on Dummy Variables **

do we need this? perhaps not.

```python
pca = PCA(n_components = 30)
df1_dummy_pca = pca.fit_transform(df1_dummy)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(pca.explained_variance_ratio_)
plt.title("Explained Variance Ratio vs. Number of PCA Components")
plt.subplot(1,2,2)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title("Accumulative Explained Variance Ratio vs. Number of PCA Components")
```

```python
# based on the explained variance ratio chart, we will use 10 PCA components
# which will explain almost 80% of the variances.
pca2 = PCA(n_components = 10)
df_dummy_pca2 = pca2.fit_transform(df1_dummy)
df_dummy_pca2[:5]
# combine numercial variables and 10 pca components as features, feature union?
# features = num_cols + df_dummy_pca2
```

## Build Models and Tuning Parameter

Fit different models, linear regression, XGB, ... and then ensemble model.

** Split Data Sets **

```python
# Partition the dataset in train + validation sets
y = df1['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(df2, y, test_size = 0.3, random_state = 0)
print("X_train : " + str(X_train.shape))
print("X_test : " + str(X_test.shape))
print("y_train : " + str(y_train.shape))
print("y_test : " + str(y_test.shape))
```

```python
# Standardize numerical features
# shall we standardize SalePrice too?
# stdSc = StandardScaler()
# X_train.loc[:, num_cols] = stdSc.fit_transform(X_train.loc[:, num_cols])
# X_test.loc[:, num_cols] = stdSc.transform(X_test.loc[:, num_cols])
```

** Linear Regression **

```python
# fit a linear regression model--don't need to fit a normal linear regression but need the residual plots
lr = LinearRegression()
lr.fit(X_train, y_train)
test_pred = lr.predict(X_test)
train_pred = lr.predict(X_train)
print('MSE train: %.3f, test: %.3f' %
     (mean_squared_error(y_train, train_pred), mean_squared_error(y_test, test_pred)))
print('R^2 train: %.3f, test: %.3f' % 
     (r2_score(y_train, train_pred), r2_score(y_test, test_pred)))

# residual plots
plt.scatter(train_pred, train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(test_pred, test_pred - y_test, c='green', marker='s', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc = 'upper right')
plt.xlim([-2500, 600000])
plt.show()
```

```python
# regularized linear regression
# Ridge regression (L2 regulation)
from sklearn.linear_model import Ridge
alphas = [0.05, 0.1, 0.3, 1, 3, 5, 10, 15, 30, 50, 75] # use Ryan's pipeline to select hyper parameter
rdg = Ridge(alpha=10.0)  # alpha is regularization strength
rdg.fit(X_train, y_train)
test_pred_rdg = rdg.predict(X_test)
train_pred_rdg = rdg.predict(X_train)
print('R^2 train: %.3f, test: %.3f' % 
     (r2_score(y_train, train_pred_rdg), r2_score(y_test, test_pred_rdg)))
```

```python
# LASSO regression
# L1 regulation, useful for feature selection
from sklearn.linear_model import Lasso
lso = Lasso(alpha=0.0001) # alpha is quite import here
lso.fit(X_train, y_train)
test_pred_lso = lso.predict(X_test)
train_pred_lso = lso.predict(X_train)
print('R^2 train: %.3f, test: %.3f' % 
     (r2_score(y_train, train_pred_lso), r2_score(y_test, test_pred_lso)))
coef = pd.Series(lso.coef_, index = X_train.columns)
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " 
      +  str(sum(coef == 0)) + " variables")
```

```python
# check selected features by LASSO
imp_coef = pd.concat([coef.sort_values().head(10),
                     coef.sort_values().tail(10)])
imp_coef.plot(kind = "barh")
plt.title("Coefficients in the Lasso Model")
```

** Other Trees **

1. sklearn.ensemble.AdaBoostRegressor
2. Gradient Boosting regression
http://scikit-learn.org/stable/auto_examples/ensemble/plot_gradient_boosting_regression.html

```python
# Gradient Boosting Regression, GBRT
from sklearn.ensemble import GradientBoostingRegressor


params = {'n_estimators': 400, 'max_depth': 4, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'} # least squares loss 
clf = GradientBoostingRegressor(**params)
clf.fit(X_train, y_train)

test_pred_gb = clf.predict(X_test)
train_pred_gb = clf.predict(X_train)
print('R^2 train: %.3f, test: %.3f' % 
     (r2_score(y_train, train_pred_gb), r2_score(y_test, test_pred_gb)))
print('MSE train: %.3f, test: %.3f' %
     (mean_squared_error(y_train, train_pred_gb), mean_squared_error(y_test, test_pred_gb)))

# #############################################################################
# Plot training deviance

# compute test set deviance
test_score = np.zeros((params['n_estimators'],), dtype=np.float64)

for i, y_pred in enumerate(clf.staged_predict(X_test)):
    test_score[i] = clf.loss_(y_test, y_pred)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, clf.train_score_, 'b-',
         label='Training Set Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, test_score, 'r-',
         label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Deviance')

# #############################################################################
# Plot feature importance
feature_importance = clf.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, X_train.columns[sorted_idx])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```

** Ensemble **

use several model prediction to fit a linear regression model for the ensemble model instead of weighted average of several model output.
1. regularized linear regression
2. Random Forecast or Ada boost tree or XGB tree
3. SVM

```python
ens = LinearRegression()
X_ens_train = pd.concat([pd.Series(train_pred_rdg),pd.Series(train_pred_forest),pd.Series(train_pred_svm)], axis=1)
X_ens_test = pd.concat([pd.Series(test_pred_rdg),pd.Series(test_pred_forest),pd.Series(test_pred_svm)], axis=1)
# print(train_pred_rdg.shape, X_ens_train.shape,X_train.shape)
ens.fit(X_ens_train, y_train)
test_pred_ens = ens.predict(X_ens_test)
train_pred_ens = ens.predict(X_ens_train)
print('MSE train: %.3f, test: %.3f' %
     (mean_squared_error(y_train, train_pred_ens), mean_squared_error(y_test, test_pred_ens)))
print('R^2 train: %.3f, test: %.3f' % 
     (r2_score(y_train, train_pred_ens), r2_score(y_test, test_pred_ens)))
```
