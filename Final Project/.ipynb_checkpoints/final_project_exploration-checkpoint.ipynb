### Multivariate Analysis of the Categorical Features

```python
string_columns = np.array(train.columns[train.dtypes == object])
num_unique = np.array([len(train[col].unique()) for col in string_columns])

print('Unique values of each categorical feature:\n')
for col, nunique in zip(string_columns, num_unique):
    print('{col} : {unique_values}\n'.format(col=col, unique_values=train[col].unique().tolist()))
```

```python
train[string_columns] = train[string_columns].fillna('Missing')
```

```python
string_columns[num_unique > 10]
```

```python
short_categorical_variables = string_columns[num_unique < 10]
short_categorical_variables = np.array(short_categorical_variables).reshape(10,4)
```

```python
# Rescale the SalePrice varaible for this plot
train['log_SalePrice'] = np.log(train['SalePrice'])
train['log_SalePrice_rescaled'] = (train['log_SalePrice'] - train['log_SalePrice'].mean()) / train['log_SalePrice'].std()

# Create a 10x4 grid of subplots 16 inches wide, 40 inches tall
f, axs = sns.plt.subplots(*short_categorical_variables.shape)
f.set_size_inches(16,40)

# Iterate through the boxplot var
for row in range(short_categorical_variables.shape[0]):
    for col in range(short_categorical_variables.shape[1]):
        sns.boxplot(ax=axs[row, col]
                    ,data=train
                    ,x=short_categorical_variables[row, col]
                    ,y='log_SalePrice_rescaled'
                   )
        # Rotate the xtick labels by 45 degrees so they don't overlap each other
        sns.plt.setp(axs[row, col].get_xticklabels(), rotation=45, fontsize=12)
        sns.plt.setp(axs[row, col].get_yticklabels(),  fontsize=12)
        
        # Remove the x and y labels (too much chart junk, but set the variable name as the title
        axs[row, col].set_xlabel('')
        axs[row, col].set_title(short_categorical_variables[row, col], fontsize=12)
        axs[row, col].set_ylabel('')

f.tight_layout()  # use the tight layout option so the subplots are nicely spaced
```

```python
long_categorical_variables = string_columns[num_unique > 10]
long_categorical_variables = np.append(long_categorical_variables, ['MSSubClass', 'OverallQual', 'OverallCond'])

f, axs = sns.plt.subplots(len(long_categorical_variables),1)
f.set_size_inches(16,30)

for i in range(len(axs)):
    
    sns.boxplot(ax=axs[i]
                , data=train
                , x=long_categorical_variables[i]
                , y='log_SalePrice_rescaled') 
    
    
    sns.plt.setp(axs[i].get_xticklabels(), rotation=45, fontsize=12)
    sns.plt.setp(axs[i].get_yticklabels(),  fontsize=12)
    
    axs[i].set_title('{} vs Sale Price Box Plots'.format(long_categorical_variables[i]), fontsize=24)
    axs[i].set_xlabel('')
    axs[i].set_ylabel('log(Sale Price)', fontsize=18)
    
f.tight_layout()
```

### Multivariate Analysis of Continuous Features

beep boop

```python
continuous_features = train.columns[train.dtypes != object].tolist()

for col in ['Id', 'log_SalePrice', 'SalePrice', 'MSSubClass', 'OverallQual', 'OverallCond']:
    continuous_features.remove(col)

print('Continuous Features:\n')
print('\n'.join(continuous_features))
```

```python
corrmat = train[continuous_features].corr()

cg = sns.clustermap(data=corrmat, figsize=(16, 16))
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0, fontsize=16)
plt.setp(cg.ax_heatmap.xaxis.get_majorticklabels(), fontsize=16)
plt.show()
```
